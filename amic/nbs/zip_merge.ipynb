```python
base = '../../data/ibamco_grouped3.csv'
amic = '../data/amic_zips.csv'
```

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv(base)
amic = pd.read_csv(amic)
```

```python
df
```

```python
amic['zipcode'] = amic['zipcode'].apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
amic['media'] = amic['media'].apply(lambda x: str(x.replace('www.', '')) if not pd.isnull(x) else x)
amic
```

```python
df['zipcode_amic'] = np.nan
```

```python
for idx, row in df.iterrows():
    media = row['media']
    url = row['url']
    if not pd.isnull(media) and media in amic['media'].values:
        # Use .at to modify the value at the given index and column
        df.at[idx, 'zipcode_amic'] = amic[amic['media'] == media]['zipcode'].iloc[0]
        print('yes')
    elif not pd.isnull(url) and url in amic['media'].values:
        # Use .at to modify the value at the given index and column
        df.at[idx, 'zipcode_amic'] = amic[amic['media'] == url.strip()]['zipcode'].iloc[0]
        print('yes')
```

```python
df[df.zipcode_amic.notnull()]
```

```python
df.to_csv('../../data/ibamco_grouped4.csv', index=False)
```
