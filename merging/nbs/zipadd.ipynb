```python
zips = '../../data/zip/zips3.xlsx'
f = '../../data/ibamco_grouped6.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(f)
zips = pd.read_excel(zips, 'main')
```

```python
zips = zips[['codigo postal', 'lat', 'lon', 'poblacion', 'provincia', 'comunidad autonoma', 'comarca2']]
zips.rename(columns={'codigo postal': 'zip', 'comarca2': 'comarca'}, inplace=True)
```

```python
zips['zip'] = zips.zip.apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
df['zipcode'] = df.zipcode.apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
```

```python
df['lat'] = np.nan
df['lon'] = np.nan
df['poblacion'] = np.nan
df['provincia'] = np.nan
df['comunidad autonoma'] = np.nan
df['comarca'] = np.nan

for i, row in df.iterrows():
    try:
        z = zips[zips.zip == row.zipcode].iloc[0]
        df.loc[i, 'lat'] = z.lat
        df.loc[i, 'lon'] = z.lon
        df.loc[i, 'poblacion'] = z.poblacion
        df.loc[i, 'provincia'] = z.provincia
        df.loc[i, 'comunidad autonoma'] = z['comunidad autonoma']
        df.loc[i, 'comarca'] = z.comarca
    except:
        pass
```
