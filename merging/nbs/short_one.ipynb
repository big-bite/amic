```python
import pandas as pd
import numpy as np
```

```python
f = '../../data/ibamco_grouped4.csv'
```

```python
df = pd.read_csv(f)
```

```python
df
```

```python
# priority : cac > amic > ojd > amic
df['zipcode_ojd'] = df['zipcode_ojd'].apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
df['zipcode'] = np.nan

for idx, row in df.iterrows():
    if pd.notnull(row['zipcode_cac']):
        df.loc[idx, 'zipcode'] = row['zipcode_cac']
    elif pd.notnull(row['zipcode_amic']):
        df.loc[idx, 'zipcode'] = row['zipcode_amic']
    elif pd.notnull(row['zipcode_ojd']):
        df.loc[idx, 'zipcode'] = row['zipcode_ojd']
    elif pd.notnull(row['zipcode_iber']):
        df.loc[idx, 'zipcode'] = row['zipcode_iber']
    
```

```python
df.to_csv('../../data/ibamco_grouped5.csv', index=False)
```

```python
df
```

```python
df = df[['ID', 'Origin', 'media', 'url', 'zipcode']]
df
```

```python
df_extend = pd.DataFrame(columns=['ID', 'Origin', 'media', 'url', 'zipcode'])
for idx, row in df.iterrows():
    if pd.notnull(row['zipcode']):
        zips = str(row['zipcode']).split(',')
        for z in zips:
            df_extend.loc[len(df_extend)] = [row['ID'], row['Origin'], row['media'], row['url'], z]
    else:
        df_extend.loc[len(df_extend)] = [row['ID'], row['Origin'], row['media'], row['url'], np.nan]
df_extend
```

```python
df_extend['zipcode'] = df_extend['zipcode'].apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
df_extend
```

```python
df_extend.to_csv('../../data/ibamco_grouped6.csv', index=False)
```
