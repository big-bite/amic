```python
import pandas as pd
import numpy as np
```

```python
file = "../data/zip/zips2.xlsx"
```

```python
df = pd.read_excel(file, 'main')
df['codigo postal'] = df['codigo postal'].astype('str').str.zfill(5)
df['code2'] = df['codigo postal'].astype('str').str.zfill(5)
df = df[['codigo postal', 'comarca2', 'provincia', 'comunidad autonoma']].set_index('codigo postal')
```

```python
ibamco = pd.read_csv('../data/ibamco_grouped2.csv')
ibamco
```

```python
new_ib = pd.DataFrame(columns=['ID', 'Origin', 'media', 'url', 'zipcode', 'comarca', 'provincia', 'comunidad autonoma'])
```

```python
df.loc['08500']
```

```python
for idx, row in ibamco.iterrows():
    zip1 = row['zipcode_cac']
    if pd.isna(zip1):
        zip1 = row['zipcode_iber']
        if pd.isna(zip1):
            zip1 = np.nan
    
    if zip1 is not np.nan:
        zips = str(zip1).split(',')
        for z in zips:
            z = z.strip()
            if z in df.index:
                new_row = [row['ID'], row['Origin'], row['media'], row['url'], z, df.loc[z]['comarca2'], df.loc[z]['provincia'], df.loc[z]['comunidad autonoma']]
            else:
                new_row = [row['ID'], row['Origin'], row['media'], row['url'], z, np.nan, np.nan, np.nan]
            new_ib.loc[len(new_ib)] = new_row
                
    else:
        new_row = [row['ID'], row['Origin'], row['media'], row['url'], np.nan, np.nan, np.nan, np.nan]
        new_ib.loc[len(new_ib)] = new_row
```
