```python
import numpy as np
import pandas as pd
```

```python
f = '../data/ojd_visits_data.csv'
```

```python
df = pd.read_csv(f)
df['zipcode'] = df['zipcode'].apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
```

```python
'''"visits_data[0] = {'browser_uniques': {'1_Month': '2.469', 'daily_avg': '93', 'mon_fri_avg': '101', 'sat_sun_avg': '70', 'pages_per_visit': '1,59', 'avg_visit_duration': '0:01:28', 'avg_interaction_time': '0:00:36'}, 'visits': {'1_Month': '3.126', 'daily_avg': '101', 'mon_fri_avg': '110', 'sat_sun_avg': '75'}, 'pages': {'1_Month': '4.962', 'daily_avg': '160', 'mon_fri_avg': '176', 'sat_sun_avg': '114'}}"'''

browser_uniques = pd.DataFrame(columns=['media', 'zipcode', '1_Month', 'daily_avg', 'mon_fri_avg', 'sat_sun_avg', 'pages_per_visit', 'avg_visit_duration', 'avg_interaction_time'])
visits = pd.DataFrame(columns=['media', 'zipcode', '1_Month', 'daily_avg', 'mon_fri_avg', 'sat_sun_avg'])
pages = pd.DataFrame(columns=['media', 'zipcode', '1_Month', 'daily_avg', 'mon_fri_avg', 'sat_sun_avg'])
import json
```

```python
for i in range(len(df)):
    media = df['media'][i]
    zipcode = df['zipcode'][i]
    val = json.loads(df['visits_data'][i].replace("'", "\""))
    browser_uniques.loc[i] = [media, zipcode, val['browser_uniques']['1_Month'], val['browser_uniques']['daily_avg'], val['browser_uniques']['mon_fri_avg'], val['browser_uniques']['sat_sun_avg'], val['browser_uniques']['pages_per_visit'], val['browser_uniques']['avg_visit_duration'], val['browser_uniques']['avg_interaction_time']]
    visits.loc[i] = [media, zipcode, val['visits']['1_Month'], val['visits']['daily_avg'], val['visits']['mon_fri_avg'], val['visits']['sat_sun_avg']]
    pages.loc[i] = [media, zipcode, val['pages']['1_Month'], val['pages']['daily_avg'], val['pages']['mon_fri_avg'], val['pages']['sat_sun_avg']]
    
```

```python
browser_uniques
```

```python
browser_uniques.to_csv('../data/visits/browser_uniques.csv', index=False)
visits.to_csv('../data/visits/visits.csv', index=False)
pages.to_csv('../data/visits/pages.csv', index=False)
```
