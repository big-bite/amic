```python
# l = 'https://www.ojdinteractiva.es/medios-digitales-comunidad-valenciana#'
# l = 'https://www.ojdinteractiva.es/medios-digitales-islas-baleares#'
l = 'https://www.ojdinteractiva.es/mitjans-digitals-en-catala#'
```

```python
# extract all links from the page
import requests
from bs4 import BeautifulSoup
import re

r = requests.get(l)
soup = BeautifulSoup(r.text, 'html.parser')
links = []
for link in soup.find_all('a', href=True):
    links.append(link['href'])

# filter out the links starting with 'https://www.ojdinteractiva.es/muestra_acta/medios-digitales'
links = [link for link in links if re.match(r'https://www.ojdinteractiva.es/muestra_acta/medios-digitales', link)]
```

```python
# get the data from the table tag as dataframe frm soup

import pandas as pd
table = soup.find_all('table')[-2]
df = pd.read_html(str(table), extract_links='all')[0]
```

```python
links_seq = df[11].apply(lambda x: x[1])
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: x[0])
```

```python
df[11] = links_seq  
```

```python
# set first row as header   
df.columns = df.iloc[0]
df = df[1:]
df.rename(columns={None: 'Link'}, inplace=True)
df.rename(columns={'TÍTOL': 'TITULO'}, inplace=True)
df = df[['TITULO', 'Link']]
df
```

```python
links_old = pd.read_csv('../data/links.csv')
links_old
```

```python
links_old = pd.concat([links_old, df], ignore_index=True)
links_old
```

```python
links_old[['TITULO', 'Link']].to_csv('../data/links.csv', index=False)
```
