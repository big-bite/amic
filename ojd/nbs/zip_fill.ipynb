```python
import pandas as pd
import numpy as np
```

```python
base = '../../data/ibamco_grouped2.csv'
zips = '../data/links_with_zipcodes.csv'
```

```python
df = pd.read_csv(base)
z = pd.read_csv(zips)
```

```python
df['url'] = df.url.apply(lambda x: x.strip() if type(x) == str else x)
```

```python
z['media'] = z['media'].str.strip().str.lower()
z['zipcode'] = z['zipcode'].apply(lambda x: str(int(float(x))).zfill(5) if not pd.isnull(x) else x)
z.dropna(subset=['zipcode'], inplace=True)
```

```python
z
```

```python
df['zipcode_ojd'] = np.nan
df
```

```python
df[df['url'].astype(str).str.contains("\(")]
```

```python
for idx, row in df.iterrows():
    media = row['media']
    url = row['url']
    if not pd.isnull(media) and media in z['media'].values:
        # Use .at to modify the value at the given index and column
        df.at[idx, 'zipcode_ojd'] = z[z['media'] == media]['zipcode'].iloc[0]
        print('yes')
    elif not pd.isnull(url) and url in z['media'].values:
        # Use .at to modify the value at the given index and column
        df.at[idx, 'zipcode_ojd'] = z[z['media'] == url]['zipcode'].iloc[0]
        print('yes')
```

```python
df.to_csv('../../data/ibamco_grouped3.csv', index=False)
```
